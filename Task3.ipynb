```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Initialize Dash app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Iris Dataset Dashboard", style={'textAlign': 'center'}),

    html.Label("Select X-axis feature:"),
    dcc.Dropdown(
        id='x-axis',
        options=[{'label': col, 'value': col} for col in iris.feature_names],
        value=iris.feature_names[0]
    ),

    html.Label("Select Y-axis feature:"),
    dcc.Dropdown(
        id='y-axis',
        options=[{'label': col, 'value': col} for col in iris.feature_names],
        value=iris.feature_names[1]
    ),

    dcc.Graph(id='scatter-plot')
])

# Callback to update the scatter plot
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('x-axis', 'value'),
     Input('y-axis', 'value')]
)
def update_graph(x_axis, y_axis):
    fig = px.scatter(df, x=x_axis, y=y_axis, color='species', title=f'{x_axis} vs {y_axis}')
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
